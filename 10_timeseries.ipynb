```python
# 예측 분석 : 시계열 분석 ( timeseries )
# 자기상관성을 가진 데이터 (Autocorelation)
    # lag(시차), shift(이동:차분하기위해), diff(차분:미분), resample(시간을 채움), 분 , 시간
    # 시계열 데이터는 하나의 열에서 벌어지는 일 (데이터 자체가 하나의 열)(앞 뒤로 작업)
    # 공분산은 두개의 열을 대상
    # interploate(빈 데이터를 채움), fill
# 시간 index
    # datetime , timedalta, period, interval(처음과 끝)
    # range를 전부 가지고 있음 (초기, 한계값, 증감값):list
    # arange :ndarray
    # Date_range, Timedelta_range, Period_range, Interval_range
```

```python
import matplotlib.pyplot as plt
# 주기성이 있는 데이터를 만듬 : 시계열데이터는 주기성을 가지고 있음
spacing=np.linspace(-5*np.pi, 5*np.pi,num=100)
s=pd.Series(0.7*np.random.rand(100)+0.3*np.sin(spacing))
print(s.autocorr()) # 자기상관성
# 상관계수의 값의 범위 -1~1(0:상관없음)
x=pd.plotting.autocorrelation_plot(s)   # 이상데이터 확인
# 차분을 통해 정상데이터로 변환해야함
x.plot()
```

```python
import numpy as np
import pandas as pd
df = pd.DataFrame({'rating': [90, 85, 82, 88, 94, 90, 76, 75, 87, 86],
                   'points': [25, 20, 14, 16, 27, 20, 12, 15, 14, 19],
                   'assists': [5, 7, 7, 8, 5, 7, 6, 9, 9, 5],
                   'rebounds': [11, 8, 10, 6, 6, 9, 6, 10, 10, 7]})
df
```

```python
# ols (ordinary least square)(평범한 최소 제곱법을 이용한 선형회귀)
from statsmodels.formula.api import ols     # statsmodels:파이썬에서 R의 기능을 사용할 수 있게 만드는 패키
model=ols('rating~ points + assists+ rebounds',data=df).fit()
print(model.summary())
```

```python
durbin_watson(model.resid)      # 잔차를 이용해서 durbin_watson 확인
```

```python
# 시계열 데이터 분해
# pandas 에서 시계열 데이터로 구분하는 기준: 인덱스가 time 계열일 때
from statsmodels.tsa.seasonal import seasonal_decompose
df=pd.read_csv('/content/drive/MyDrive/daejeon/AirPassengers.csv')
df.head()
```

```python
df.info()
```

```python
from datetime import datetime
df["Month"]=pd.to_datetime(df['Month'],infer_datetime_format=True)
df.index=df['Month']
df
```

```python
# SVD 특이 행렬분해
# 시계열 분석은 float 숫자데이터에만
del df['Month']
decomposed = seasonal_decompose(df,model='addictive')   # 가법(더하는 ), 승법(곱해져서 만들어진 데이터 일 것이다)모델(multipliactive)
```

```python
# multipliactive,addictive 둘 다 적용 후 선택
decomposed = seasonal_decompose(df,model='multipliactive')
```

```python
decomposed
```

```python
trend=decomposed.trend  # 추세
seasonal=decomposed.seasonal    #계절변동(주기성)
residual=decomposed.resid   #잔차
```

```python
plt.plot(df['AirPassengers'],label='original')
plt.plot(trend,label='trend')
plt.plot(seasonal, label='seasonal')
plt.plot(residual,label='residual')
plt.legend()
plt.show()
```
