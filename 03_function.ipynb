```python
# 재귀함수
def factorial(n):
  sum=1         # 1이 포함이 안돼서 1로
  for i in range (n,1,-1):
    sum *= i
  return sum
factorial(5)
```

```python
# 재귀함수 : 자기 자신을 호출
def factorial(n):
  if n<=1:     # 종료 조건
    return 1
  return n*factorial(n-1)
factorial(5)
```

```python
# 재귀호출 : binary search
# 전제조건 :정렬된 데이터를
name_list=['진정인','원하는데로','희망이','협동인',
           '통일대박','딥러닝','다함께','대한인']
name_list.sort()           # 오름차순
print(name_list)
def binary_search_recursive(search_list, key, lower_bound, upper_bound):
  if lower_bound>upper_bound:
    print('찾는 데이터 없음')
    return
  middle_pos=(lower_bound+upper_bound)//2     # 중앙지점에서 비교
  if middle_pos <=0 or middle_pos >= len(search_list):  # 교차되었다는 얘기, 데이터가 없다
    print("찾는 데이터 없음")
    return
  if search_list[middle_pos] < key:   # 정렬되어 있음, 데이터가 크다, 중앙에서 오른쪽에 있다
    binary_search_recursive(search_list, key, middle_pos+1, upper_bound)  # middle_pos=lower_bound, 중간값은 아니니까 +1 해줬음
  elif search_list[middle_pos]> key:
    binary_search_recursive(search_list, key, lower_bound, middle_pos)
  else:
    print('찾는 데이터가 있는 인덱스', middle_pos)
lower_bound=0
upper_bound=len(name_list)
search_word=input('검색하고자 하는 단어를 입력')
binary_search_recursive(name_list, search_word, lower_bound,upper_bound)
```

```python
# 문제
# 두 점의 사이각을 구하는 함수를 작성하시오
# 요소끼리 곱해서 다 더해주면댐
# costheta= 내적한결과/norm1*norm2
# radian=acos(costheta)
# math.degree
import math
def theta_netween_points(x1,y1,x2,y2):
  dot=x1*x2+y1*y2
  norm1=math.sqrt(x1**2+y1**2)
  norm2=math.sqrt(x2**2+y2**2)
  costheta=dot/(norm1*norm2)
  thetaradians=math.acos(costheta)
  theta = math.degrees(thetaradians)
  return theta

a=0,1
b=1,1
theta_netween_points(*a,*b)
```
