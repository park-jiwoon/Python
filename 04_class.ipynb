```python
# class method
from datetime import date
class Person:
  def __init__(self,name,age):
    self.name=name
    self.age=age
  @classmethod  # 다양한 초기화 방법이 필요한 경우 # 클래스 자체에 영향을 주고 싶은 경우
  def fromBirthYear(cls,name,birthyear):
    return cls(name, date.today().year-birthyear)     # cls 로 생성자 호출     @출생연도만 쓴사람의 나이도 계산하고 싶은 경우
  def display(self):
    print(self.name+"의 나이는 : " +str(self.age))

person=Person('강감찬',19)
person.display()
person1=Person.fromBirthYear("을지문덕",1985)
person1.display()
```

```python
import random
class LottoBall:
  def __init__(self,ballnum):
    self.ballnum=ballnum          # 로또 번호 하나 저장
  def __repr__ (self):
    return str(self.ballnum)      # 번호 출력

class LottoMachine:  #            # 여러개의 로또 볼
  def __init__(self):
    self.ballList=[]
    for i in range(1,46):
      self.ballList.append(LottoBall(i))
  def selectBalls(self):
   random.shuffle(self.ballList);  # 섞은 후
   return self.ballList[0:6]       # 6개 리턴

class LottoUI:                     # 행동 툴
  def __init__ (self):
    self.machine=LottoMachine()
  def playLotto(self):
    input("로또를 뽑을까요?");
    selectedBalls=self.machine.selectBalls()
    for ball in selectedBalls:
      print(ball,end=' ')

lo=LottoUI()
lo.playLotto()
```

# Point class 를 작성하시오
-x,y 좌표를 데이터로 함  2차원

-값을 변경할 수 있는 get, set 함수를 작성

_두개의 포인트간의 거리값을 구하는 함수를 작성

-두개의 거리값이 3보다 작으면 충돌처리 결과를 리턴하는 함수를 작성

-충돌함수를 이용해서 결과를 리턴해주는 함수를 작성

```python
  # x,y
import random
import math
class Point:  # 클래스 정의
    def __init__(self, x, y):  # 메서드 정의
        self.x = x
        self.y = y
    def __repr__(self):
      return "x좌표:"+str(self.x)+" y:"+ str(self.y)      # 출력
    def getx(self):
      return self.x
    def gety(self):
      return self.y
    def setx(self,x):
      self.x=x
    def sety(self,y):
      self.y=y
    def dist(self,other): # python은 데이터 타입 지정x
      dx=self.x-other.x
      dy=self.y-other.y
      distance=math.sqrt(dx**2+dy**2)
      return distance
    def collision(self, other):
        return self.dist(other)<3
    def collisionUI(self, other):
        if self.collision(other):
          print("충돌!")
        else:
          print("충돌하지 않음")

first=Point(7,3)
second=Point(4,7)
print(first)
print(second)
print("두 점사이의 거리는 ", str(first.dist(second)))
if first.collision(second):
   print("충동")
else:
   print("충돌x")
first.collisionUI(second)
```

```python
# 문제
# 위의 포인트를 리스트에 10개를 인스턴스해서 입력합니다
# 이 때 좌표점은 random으로 합니다 (random.randint(1,21))
# 10개의 좌표간에 몇 개가 충돌하는지 확인하시오
import random
pointlist=[]      # 값을 저장해둘 리스트가 필요
for count in range(10):
  po=Point(random.randint(0,21),random.randint(0,21))
  pointlist.append(po)
print(count) # 0~9
totalcollisionscount=0
for count1 in range(0,9):
  for count2 in range(count1+1,10):
    if (pointlist[count1].collision(pointlist[count2])):
      print("충돌~!~!")
      print("인덱스="+str(count1),"-",str(count2))
      print("좌표점=",pointlist[count1],'-',str(pointlist[count2]))
      totalcollisionscount+=1
print("전체 충돌 횟수:"+str(totalcollisionscount))
```
