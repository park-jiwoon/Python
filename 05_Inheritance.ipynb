```python
# 부모가 자식을 가리킬 수 있다(delegate)
class Guard:
  def three_point(self):
    print("가드는 3점슛 장착")
  def apass(self):
    print("가드의 핵심은 에이패스지")
class Jordan:
  def three_point(self):
    print("조단은 3점슛 장착")
  def apass(self):
    print("조단은 에이패스 가능")
# 자바에서는 상속이 아니면 대입이 불가능
# 상속이 없어도 하나의 매개변수에 다 입력이 가능

def inhecourt(duck):    # 매개변수의 데이터타입
  duck.three_point();
  duck.apass()
gu=Guard()
jo=Jordan()
inhecourt(gu)
inhecourt(jo)
# 재정의는 남아있고, 이걸 이용->design pattern을 만들어 사용
# 재정의가 남아있는 이유:약속을 만들기 위해(오타 등을 방지하기 위해)
```

```python
# 벡터의 요소연산(때상이 2개)
# 파이선에서는 오버로딩이 불가 => 연산자 오버라이딩으로 해결
class Vector:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def __add__(self,other): # 오버라이딩 (object)
    return Vector(self.x+other.x,self.y+other.y)      # 원래 벡터, 다른 xy가진 벡터, 입력된벡터
  def __mul__(self,other):
    return Vector(self.x*other.x,self.y*other.y)
  def __sub__(self,other):
    return Vector(self.x-other.x,self.y-other.y)
  def __truediv__(self,other):  # 부동소수점 , __floorciv__ 정수
    return Vector(self.x/other.x,self.y/other.y)
  def __repr__(self):
    return "(좌표는 %d, %d )" % (self.x,self.y)
```

```python
#  문제 : 벡터
# 단일 벡터
# 원으로부터의 크기를 나타내는 norm함수를 구현
# 원점으로부터 각도를 나타내는 normamlize를 구현
# 두개의 벡터
# 두 벡터의 거리값을 구하는 함수 dist 구현
# 두 벡터의 내적, costheta, theta 를 구하는 함수를 구현
import math
class Vector:
  def __init__(self,x,y):
    self.x=x
    self.y=y
  def __add__(self,other): # 오버라이딩 (object)
    return Vector(self.x+other.x,self.y+other.y)      # 원래 벡터, 다른 xy가진 벡터, 입력된벡터
  def __mul__(self,other):
    return Vector(self.x*other.x,self.y*other.y)
  def __sub__(self,other):
    return Vector(self.x-other.x,self.y-other.y)
  def __truediv__(self,other):  # 부동소수점 , __floorciv__ 정수
    return Vector(self.x/other.x,self.y/other.y)
  def __repr__(self):
    return "(좌표는 %d, %d )" % (self.x,self.y)
  def norm(self):
    return math.sqrt(self.x**2+self.y**2)
  def normalize(self):
    norm=self.norm()
    return self.x/norm,self.y/norm
  def dist(self,other):
    return (math.sqrt((other.x-self.x)**2+(other.y-self.y)**2) )
  def dot(self,other):
    return (self.x*other.x+self.y*other.y)
  def costheta(self,other):
    dot=self.dot(other)
    norm1=self.norm()
    norm2=other.norm()
    return dot/(norm1*norm2)
  def theta(self,other):
    costheta=self.costheta(other)
    radian =math.acos(costheta)
    angle=math.degrees(radian)
    return angle
  def __repr__(self):
    return f"x좌표:{self.x}, y좌표:{self.y}"

```

```python
f=Vector(10,0)
g=Vector(0,27)
print(f)
print(f+g)
print("f벡터의 크기",f.norm())
print("g벡터의 크기",g.norm())
print("f의 normalize:",f.normalize())
print("g의 normalize:",g.normalize())
print("두 벡터의 내적=",f.dot(g))
print("두 벡터의 코사인 사이각=",f.costheta(g))
print("두 벡터의 사이각=",f.theta(g))
```

```python
class Student:
    def __init__(self, id, name, korea, english, math):
        self.id = id
        self.name = name
        self.korea = korea
        self.english = english
        self.math = math

    def average(self):
        return (int(self.korea) + int(self.english) + int(self.math)) / 3

    def __repr__(self):
        return f"{self.id} (이름: {self.name}, 국어: {self.korea}, 영어: {self.english}, 수학: {self.math})"

class CRUD:
    def __init__(self):
        self.students = {}

    def create(self):
        student_id = input("학번 입력: ")
        name = input("이름 입력: ")
        korea = input("국어 점수 입력: ")
        english = input("영어 점수 입력: ")
        math = input("수학 점수 입력: ")
        student = Student(student_id, name, korea, english, math)
        if student_id in self.students:
            print(f"학번 {student_id}은(는) 이미 존재하는 학생입니다. 추가할 수 없습니다.")
        else:
            self.students[student_id] = student
            print(f"{name} 학생이 추가되었습니다.")
        return

    def read(self):
        student_id = input("학번 입력: ")
        if student_id in self.students:
            print(self.students[student_id])
        else:
            print("찾는 학번이 없습니다.")
        return

    def update(self):
        student_id = input("학번 입력: ")
        if student_id in self.students:
            student = self.students[student_id]
            keys = input("이름 : 0, 국어 : 1, 영어 : 2, 수학 : 3")
            if keys == "0":
                b = input("수정할 이름을 입력해주세요")
                student.name = b
            elif keys == "1":
                c = input("수정할 국어점수를 입력하세요")
                student.korea = c
            elif keys == "2":
                d = input("수정할 영어점수를 입력하세요")
                student.english = d
            elif keys == "3":
                e = input("수정할 수학점수를 입력하세요")
                student.math = e
        else:
            print("수정할 학번이 없습니다.")
        return

    def delete(self):
        student_id = input("학번 입력: ")
        if student_id in self.students:
            del self.students[student_id]
            print("삭제완료", self.students)
        else:
            print("학번이 없습니다.")
        return

    def allread(self):
        if not self.students:
            print("등록된 학생이 없습니다.")
        else:
            for student_id, student in self.students.items():
                print(student)
        return

    def alldel(self):
        if not self.students:
            print("등록된 학생이 없습니다.")
        else:
            self.students.clear()
            print("졸업하여 데이터가 삭제되었습니다.")
        return

    def rank(self):
        if not self.students:
            print("등록된 학생이 없습니다.")
        else:
            sorted_students = sorted(self.students.values(), key=lambda x: x.average(), reverse=True)
            print("평균 점수를 기준으로 순위:")
            for rank, student in enumerate(sorted_students, start=1):
                print(f"순위 {rank}: {student}, 평균 점수: {student.average():.2f}")

    def aborad(self):
        if not self.students:
            print("등록된 학생이 없습니다.")
        else:
            print("어학연수 가능한 학생 목록:")
            for student_id, student in self.students.items():
                if student.average() >= 190:
                    print(student)
            if all(student.average() < 190 for student in self.students.values()):
                print("어학연수 가능한 학생이 없습니다.")
        return

class stuUI:
    crud = CRUD()
    menu = ["학생추가(1), 검색(2), 정보수정(3), 삭제(4), 전체학생조회(5), 등수확인(6), 유학가능(7), 졸업(8), 종료(9)"]
    while 1:
        stu = input(menu)
        if stu == "1": crud.create(); continue
        if stu == "2": crud.read(); continue
        if stu == "3": crud.update(); continue
        if stu == "4": crud.delete(); continue
        if stu == "5": crud.allread(); continue
        if stu == "6": crud.rank(); continue
        if stu == "7": crud.aborad(); continue
        if stu == "8": crud.alldel(); continue
        if stu == "9": break
        else: print("번호를 확인해 주세요"); continue
```
